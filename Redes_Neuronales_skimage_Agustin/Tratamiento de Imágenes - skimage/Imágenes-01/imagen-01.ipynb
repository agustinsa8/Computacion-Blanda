## Cargar y visualizar una imagen

_El Módulo skimage contiene algoritmos de tratamiento de imágenes. Las clases imread e imshow del módulo, permiten subir y visualzar la imagen, respectivamente_
_El método shape permite calcular las dimensiones del objeto. La respuesta contiene tres valores: anchura, altura y canales. Este último valor hace referencia al despliegue de la imagen (transparencia)_

```python
from skimage import io
img = io.imread('puppy.jpg')
io.imshow(img)
img.shape
```
